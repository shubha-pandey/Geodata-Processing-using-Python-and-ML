### Applying pseudiocolor schemes to image plots

Pseudocolor can be a useful tool for embracing contrast and visualizing data more easily.
Pseudocolor is only relevant to single-channel, grayscale.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread(r'D:\New folder (4)\IIRS\kashi.png')

plt.title("Composite Image")
plt.imshow(img)
```

```python
img.shape
```

```python
# extract red band

red_data = img[:,:,0]

red_data
```

```python
plt.title("Single Band Image (Red)")
plt.imshow(red_data)
```

```python
plt.title("Single Band Image (Red), Gray Color Map")
plt.imshow(red_data, cmap='gray')
```

```python
plt.close()
```
